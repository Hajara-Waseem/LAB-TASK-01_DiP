```python
from PIL import Image
import matplotlib.pyplot as plt

# Assuming the uploaded image is 'bowsprit_daisy_1_559463 (1).jpg'
image_path = list(uploaded.keys())[0]

# Open the image
img = Image.open(image_path)

# Split the image into its RGB components
r, g, b = img.split()

# Display each component
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(r, cmap='Reds')
plt.title('Red Channel')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(g, cmap='Greens')
plt.title('Green Channel')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(b, cmap='Blues')
plt.title('Blue Channel')
plt.axis('off')

plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Assuming the uploaded image path is still available from the previous step
image_path = list(uploaded.keys())[0]

# Open the image
img = Image.open(image_path)

# Convert the image to grayscale first
grayscale_img = img.convert('L')

# Convert to binary image using a threshold
# Pixels above the threshold become white (255), and below become black (0)
# You might need to adjust the threshold value (e.g., 128) based on your image.
threshold = 128
binary_img = grayscale_img.point(lambda p: 255 if p > threshold else 0)

# Display the binary image
plt.figure(figsize=(7, 7))
plt.imshow(binary_img, cmap='gray')
plt.title('Binary Image')
plt.axis('off')
plt.show()
```
